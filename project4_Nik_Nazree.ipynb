## Capstone Project: Targeted Financial Assistance to Entrepreneurs

## Introduction

In general, successful entrepreneurs can create value economically. This value allows entrepreneurs to create new products and services, stimulate new employment, which ultimately results in the acceleration of economic development. In fact, entrepreneurship is the key to economic growth where it contributes to Gross Domestic Product (GDP) via the contribution of Gross Operating Surplus (GOS). So public policies that encourages and supports entrepreneurship should be considered important for economic growth. Hence, with sustainable financial support from the government, it can further improve the performance of the businesses and contribute significantly to economic growth.

In Malaysia, although the government has allocated a large amount for financial assistance and provided various programs for small and medium businesses (SMEs), all these efforts are seen as useless when the effects of business activities are found to be insignificant. (Hung, Effendi Abdullah, Abdul Talib, & Abdul Rani, 2011). Public policies that have been established to foster business or entrepreneurship have seen limited success.

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Checking accuracy of the model
def train_test_rmse(dataTF, feature_cols):
    X = dataTF[feature_cols]
    y = dataTF.roi_sr
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
    
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    
    y_pred = linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
print(train_test_rmse(dataTF, ['state_code', 'sme_category', 'nkea_sector', 'PAT_2018', 'Disbursement_TOTAL']))
print(train_test_rmse(dataTF, ['state_code', 'nkea_sector', 'PAT_2018', 'Disbursement_TOTAL']))
print(train_test_rmse(dataTF, ['state_code', 'nkea_sector', 'Disbursement_TOTAL']))
```

```python
mse_values = []
rmse_values = []
scores = []
n = 0

print("~~~~ CROSS VALIDATION each fold ~~~~")
for train_index, test_index in kf.split(X, y):
    linreg = LinearRegression().fit(X.iloc[train_index], y.iloc[train_index])
    
    mse_values.append(metrics.mean_squared_error(y.iloc[test_index], linreg.predict(X.iloc[test_index])))
    rmse_values.append(np.sqrt(metrics.mean_squared_error(y.iloc[test_index], linreg.predict(X.iloc[test_index]))))
    scores.append(linreg.score(X, y))
    
    n += 1
    
    print('Model {}'.format(n))
    print('MSE: {}'.format(mse_values[n-1]))
    print('RMSE: {}'.format(rmse_values[n-1]))
    print('R2: {}\n'.format(scores[n-1]))


print("~~~~ SUMMARY OF CROSS VALIDATION ~~~~")
print('Mean of MSE for all folds: {}'.format(np.mean(mse_values)))
print('Mean of RMSE for all folds: {}'.format(np.mean(rmse_values)))
print('Mean of R2 for all folds: {}'.format(np.mean(scores)))
```
